## **概要**

麻雀の手牌を入力するとその手牌で和了れる牌が出力されます．

## **操作方法**

**手牌を入力する方法**

実行すると一番下に牌の名称が書かれている選択リストが表示されます．このリストは，選択している牌の欄に色が付きます．

入力したい牌を選択し，リストの下にある緑色の選択ボタンを押してください．選択ボタンを押したら赤い枠の中に選択した手牌が表示されます．

13枚の手牌を入力すると選択リストの表示が消え，青色の確定ボタンが表示されます．確定ボタンを押し，手牌を確定させると対応する待ちを調べて表示します．

**入力を間違えた場合**

選択ボタンや確定ボタンの右側にあるオレンジ色の変更ボタンを押してください．確定ボタンを押すと選択リストが数字に代わります．これは，赤い枠の中に表示されている牌の名称の上に書いている数字に対応しています．

数字を選んで選択ボタンを押すと牌の選択リストが出てくるので上記の「**手牌を入力する方法**」の通りに入力してください

```python
# @title 複合形探索 new stile
# haishi_num=[1,2,3,12,13,20,20,24,25,26,32,32,32]
form=[[],[],[],[]]


# 順子ブロック
def step(forms):
  an={}
  block=[]
  while len(forms)>2:
    for b in range(len(forms)-2):
      if forms[b]+1 in forms and forms[b]+2 in forms:
        an["type"]="step"
        an["number"]=forms[b]
        forms.remove(forms[b]+2)
        forms.remove(forms[b]+1)
        forms.remove(forms[b])
        break
    else:
      break
  return forms

# 刻子ブロック
def trio(forms):
  an={}
  block=[]
  while len(forms)>2:
    for b in range(len(forms)-2):
      if forms[b] == forms[b+1] and forms[b] == forms[b+2]:
        an["type"]="trio"
        an["number"]=forms[b]
        forms.remove(forms[b])
        forms.remove(forms[b])
        forms.remove(forms[b])
        break
    else:
      break
  return forms

# 配列の形を変更する関数
def into_form_from_mat(f:list):
  form=[]
  for i in range(len(f)):
    for j in range(f[i]):
      form.append(i+1)
  return form

# 複合形を固定した時に矛盾が発生しないか確かめる関数
def confirmer(a,b):
  forms=copy.deepcopy(a)
  for delete in b:
    forms.remove(delete)
  if len(forms) % 3 == 0:
    forms = step(forms)
    forms = trio(forms)
  elif len(forms) % 3 == 2:
    forms = step(forms)
    forms = trio(forms)
  return len(forms) == (len(a)-len(b)) % 3


# 重複アリの２つの配列の要素を比較する関数(インターネットからコピー)
def equal_list(lst1, lst2):
  lst = lst1.copy()
  for element in lst2:
    try:
      lst.remove(element)
    except ValueError:
      break
  else:
    return True
  return False


def search_comp(forms:list):
  # 複合系が存在しうる牌の枚数
  a = [1,2,4,5,7,8,10,11,13]
  waits = [0] * 9
  sendable = 0
  changing = 0
  block = []
  for to_break in range(1):
    if len(forms) in a:
      pass
    else:
      step(forms)
      trio(forms)
      break
    m = a.index(len(forms))
    for i in range(m+1):
      search_num = m-i
      if search_num > 2:    #面子を含む複合系(4枚以上)のとき
        for t in range(len(compound_dict[search_num])):
          if m%2 == 1 and search_num%2 == 0:
            break
          # 辞書から中身を取り出す
          comp_dict = copy.deepcopy(compound_dict[search_num][t])
          action=copy.deepcopy(comp_dict['forms'])
          comp_len = copy.deepcopy(comp_dict['len'])
          w = copy.deepcopy(comp_dict['waits'])
          if comp_dict['sendable'] == 1:
            for send in range(9):
              w[send]*=2
          start,stop=-1,9-comp_len
          steps=1
          if len(forms) < comp_dict['quant']:
            continue
          # 取り出した複合形があるか検索
          for reverse in range(2):  # 複合形の上下を返して2回
            start,stop = -stop,-start
            steps *= -1
            action.reverse()
            w.reverse()
            if len(forms) < comp_dict['quant']:
              break
            if comp_len >= 8:   # 8以上のときは移動がいらない
              start,stop = 0,1
            for n in range(start,stop,steps):
              if len(forms) < comp_dict['quant']:
                break
              comp_form = into_form_from_mat(copy.deepcopy(action))
              for j in range(len(comp_form)):
                comp_form[j] += n
              # 検索する関数の実行
              tf=equal_list(forms,comp_form)
              if tf:  # 複合系があったとき
                if confirmer(forms,comp_form):
                  pass
                else:
                  continue
                for delete in comp_form:
                  forms.remove(delete)
                changing += 1
                # 待ちを辞書から取り出す
                if n == -1:
                  w[:n] = w[-n:]
                else:
                  w[n:] = w[:9-n]
                for slide in range(n):
                  w[slide] = 0
                # 情報をもとに待ちを整理
                if comp_dict['sendable'] == 1:
                  sendable += 1
                for sum in range(9):
                  waits[sum] = waits[sum]+w[sum]
              for j in range(len(comp_form)):
                comp_form[j] += n
      else:   #面子を含まない複合系(2枚以下)のとき
        # 面子を取り除く
        step(forms)
        trio(forms)
        for t in range(len(compound_dict[search_num])):
          if m%2 == 1 and search_num%2 == 0:
            break
          if len(forms) < search_num:
            break
          # 辞書から中身を取り出す
          comp_dict = copy.deepcopy(compound_dict[search_num][t])
          action=copy.deepcopy(comp_dict['forms'])
          comp_len = copy.deepcopy(comp_dict['len'])
          w = copy.deepcopy(comp_dict['waits'])
          if comp_dict['sendable'] == 1:
            for send in range(9):
              w[send] *= 2
          start,stop = -1,9-comp_len
          steps = 1
          # 取り出した複合形があるか検索
          for reverse in range(2):
            start,stop = -stop,-start
            steps *= -1
            action.reverse()
            w.reverse()
            if comp_len >= 8:
              start,stop = 0,1
            for n in range(start,stop,steps):
              comp_form = into_form_from_mat(copy.deepcopy(action))
              for j in range(len(comp_form)):
                comp_form[j] += n
              # 検索する関数の実行
              tf=equal_list(forms,comp_form)
              if tf:  # 複合系があったとき
                if confirmer(forms,comp_form):
                  pass
                else:
                  continue
                for delete in comp_form:
                  forms.remove(delete)
                changing += 1
                # 待ちを辞書から取り出す
                if n == -1:
                  w[:n] = w[-n:]
                else:
                  w[n:] = w[:9-n]
                for slide in range(n):
                  w[slide] = 0
                # 情報をもとに待ちを整理
                if comp_dict['sendable'] == 1:
                  sendable += 1
                for sum in range(9):
                  waits[sum] = waits[sum]+w[sum]
              for j in range(len(comp_form)):
                comp_form[j] += n
  return waits,sendable,changing
```
